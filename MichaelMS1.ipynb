```python
from herbie import Herbie
import pandas as pd
import os
from herbie import Herbie, FastHerbie
import pandas as pd, numpy as np
import xarray as xr
import dask

# -----------------------------
# SETTINGS
# -----------------------------
dt = pd.Timestamp("2026-03-16 18:00")
fxx = range(1, 13)   # shorter range for testing

data_dir = "href_data"

files = []

# -----------------------------
# DOWNLOAD LOOP
# -----------------------------
for f in fxx:
    try:
        H = Herbie(
            dt,
            model="ifs",          # 🔥 SWITCHED MODEL
            product="oper",         # surface fields
            fxx=f,
            save_dir=data_dir
        )

        file = H.download()

        if file and os.path.exists(file):
            files.append(file)
            print(f"✅ f{f:02d} downloaded")
        else:
            print(f"❌ f{f:02d} missing")

    except Exception as e:
        print(f"⛔ f{f:02d} error:", e)

print(f"\n📦 Total files: {len(files)}")
```

```python
valid_files = [
    f for f in files
    if os.path.exists(f) and os.path.getsize(f) > 10000
]

print(f"✅ Valid files: {len(valid_files)}")
```
