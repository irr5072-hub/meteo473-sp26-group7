```python
import math

# Function to calculate solar radiation
def calculate_solar_radiation(n, phi, delta):
    """
    Calculate the solar radiation (S) at solar noon for a given day of the year.
    
    :param n: Day of the year (1 to 365)
    :param phi: Latitude in degrees (e.g., 40.7934 for State College, PA)
    :param delta: Solar declination angle in degrees
    
    :return: Solar radiation (S) in W/m²
    """
    S_o = 1367  # Solar constant in W/m²
    
    # Solar declination angle adjustment for Earth's elliptical orbit
    correction = 1 + 0.034 * math.cos(2 * math.pi * (n - 1) / 365)
    
    # Calculate solar zenith angle at solar noon (simplified)
    theta = math.acos(math.sin(math.radians(phi)) * math.sin(math.radians(delta)) +
                      math.cos(math.radians(phi)) * math.cos(math.radians(delta)))
    
    # Solar radiation (S) at surface
    S = S_o * math.cos(theta) * correction
    return S

# Function to calculate WBGT
def calculate_wbgt(T_db, T_wb, T_g, S, method='outdoor'):
    """
    Calculate the Wet Bulb Globe Temperature (WBGT) based on the provided temperatures and solar radiation.
    
    :param T_db: Dry bulb temperature in degrees Celsius
    :param T_wb: Wet bulb temperature in degrees Celsius
    :param T_g: Globe temperature in degrees Celsius
    :param S: Solar radiation in W/m²
    :param method: 'outdoor' or 'indoor' method to calculate WBGT
    :return: WBGT in degrees Celsius
    """
    if method == 'outdoor':
        # Formula for outdoor conditions (with solar radiation and direct sunlight)
        wbgt = 0.7 * T_wb + 0.2 * T_g + 0.1 * T_db
    else:
        # Formula for indoor conditions (no sunlight)
        wbgt = 0.7 * T_wb + 0.3 * T_db
    
    return wbgt

# Data for each date
dates = {
    "January 15th": {
        "n": 15,
        "delta": -21.26,  # Solar declination for January 15th
        "T_db": 20,  # Example Dry Bulb Temperature in °C
        "T_wb": 18,  # Example Wet Bulb Temperature in °C
        "T_g": 25    # Example Globe Temperature in °C
    },
    "March 15th": {
        "n": 74,
        "delta": -9.91,  # Solar declination for March 15th
        "T_db": 20,  # Example Dry Bulb Temperature in °C
        "T_wb": 18,  # Example Wet Bulb Temperature in °C
        "T_g": 25    # Example Globe Temperature in °C
    },
    "April 15th": {
        "n": 105,
        "delta": 9.47,  # Solar declination for April 15th
        "T_db": 20,  # Example Dry Bulb Temperature in °C
        "T_wb": 18,  # Example Wet Bulb Temperature in °C
        "T_g": 25    # Example Globe Temperature in °C
    }
}

# Latitude for State College, PA
phi = 40.7934  # Latitude in degrees

# Loop through each date and calculate solar radiation and WBGT
for date, values in dates.items():
    # Get the day of the year and other values
    n = values["n"]
    delta = values["delta"]
    T_db = values["T_db"]
    T_wb = values["T_wb"]
    T_g = values["T_g"]
    
    # Calculate solar radiation for this date
    S = calculate_solar_radiation(n, phi, delta)
    
    # Calculate WBGT for outdoor conditions
    wbgt = calculate_wbgt(T_db, T_wb, T_g, S, method='outdoor')
    
    # Print the results
    print(f"Date: {date}")
    print(f"Estimated Solar Radiation: {S:.2f} W/m²")
    print(f"Estimated WBGT: {wbgt:.2f} °C\n")
```
